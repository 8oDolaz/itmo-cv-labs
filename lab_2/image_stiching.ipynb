```python
import os

import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

```python
def plot_image(img):
    img = cv.cvtColor(img, cv.COLOR_BGR2RGB)

    plt.imshow(img)
    plt.axis('off')
    plt.show()
```

```python
base_path = os.path.join(os.getcwd(), 'panorama_images')
all_images_path = sorted(os.listdir(base_path))
all_images_path = [os.path.join(base_path, img) for img in all_images_path]

images = np.array([
    cv.imread(img_path)
    for img_path in all_images_path
])

plot_image(np.concatenate(images, axis=1))
```

```python
sticher = cv.Stitcher.create(cv.Stitcher_PANORAMA)

status, panorama_image = sticher.stitch(images)

plot_image(panorama_image)
```
